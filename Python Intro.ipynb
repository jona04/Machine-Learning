```python
import pandas 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('lsd_math_score_data.csv')
```

```python
data
```

```python
time = data[['Time_Delay_in_Minutes']]
LSD = data[['LSD_ppm']]
score = data[['Avg_Math_Test_Score']]
```

```python
%matplotlib inline

plt.title('Tissue concentration of LSD over time', fontsize=17)
plt.xlabel('Time in Minutes')
plt.ylabel('LSD concentration in PPM')
plt.text(x=0,y=-0.5, s='Wagner et al. (1968)')
plt.xticks(fontsize=14)

plt.ylim(1,7)
plt.xlim(0,500)

plt.style.use('classic')

plt.plot(time,LSD, color='g', linewidth=3)
plt.show()
```

```python
regr = LinearRegression()
regr.fit(LSD,score)
print('Theta1 ',regr.coef_[0][0])
print('Theta2 ',regr.intercept_[0])
print('R-square ',regr.score(LSD,score))
predicted_score = regr.predict(LSD)
```

```python
%matplotlib inline

plt.title('Arithmetic vc LSD-25')
plt.xlabel('Tissue LSD ppm')
plt.ylabel('Performance Score')
plt.ylim(25,85)
plt.xlim(1,6.5)

plt.style.use('fivethirtyeight')

plt.scatter(LSD,score, color='blue', s=100)
plt.plot(LSD,predicted_score, color='red')
plt.show()
```
